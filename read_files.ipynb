# JSON ва аз API хондан

Барои хондани JSON файлҳо ба мо аввал китобхонаи requests лозим аст. Ва баъд link - ро гирифта истифода бурда метавонем

```python
import requests
r = requests.get('http://openlibrary.org/subjects/python.json?published_in=2010-2024')
```

```python
r.status_code # Агар ба ҷои 200 дигар ададҳо масалан 404 барояд ин хато кор карда истодааст
```

```python
json_df = pd.DataFrame(r.json()['works']) # Тарзи хондани JSON - файл
json_df.head()
```

# Тарзи сохранит кардани JSON файл

```python
json_df.to_json('openlibrary_python.json')
```
